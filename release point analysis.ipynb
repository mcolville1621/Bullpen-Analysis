```python
import pandas as pd
import numpy as np
from pybaseball import pitching_stats
```

# Vertical Release Point Analysis

```python
RP_comparison = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Pitch Analysis\Release Points.csv")[["player_name","RP_z", "RP_x","ba", "iso", "babip", "slg", "woba", "hits", "launch_speed", "launch_angle", "spin_rate", "velocity", "whiffs"]]
RP_comparison['Type'] = np.where(RP_comparison['RP_z'] < 4.7, "Submarine", "Normal")
RP_comparison.columns
```

```python
RP_comparison['vert_quartile'] = np.where((RP_comparison['RP_z'] < RP_comparison['RP_z'].quantile(.1)), '10% less',
    np.where((RP_comparison['RP_z'] < RP_comparison['RP_z'].quantile(.25)) & (RP_comparison['RP_z'] >= RP_comparison['RP_z'].quantile(.1)),'10-25%',
    np.where((RP_comparison['RP_z'] < RP_comparison['RP_z'].quantile(.5)) & (RP_comparison['RP_z'] >= RP_comparison['RP_z'].quantile(.25)), '25-50%',
    np.where((RP_comparison['RP_z'] < RP_comparison['RP_z'].quantile(.75)) & (RP_comparison['RP_z'] >= RP_comparison['RP_z'].quantile(.5)), '50-75%',
    np.where((RP_comparison['RP_z'] < RP_comparison['RP_z'].quantile(.90)) & (RP_comparison['RP_z'] >= RP_comparison['RP_z'].quantile(.75)), '75-90%', '90% >')))))
```

```python
RP_comparison['horiz_quartile'] = np.where((RP_comparison['RP_x'] < RP_comparison['RP_x'].quantile(.1)), '10%>',
    np.where((RP_comparison['RP_x'] < RP_comparison['RP_x'].quantile(.25)) & (RP_comparison['RP_x'] >= RP_comparison['RP_x'].quantile(.1)),'10-25%',
    np.where((RP_comparison['RP_x'] < RP_comparison['RP_x'].quantile(.5)) & (RP_comparison['RP_x'] >= RP_comparison['RP_x'].quantile(.25)), '25-50%',
    np.where((RP_comparison['RP_x'] < RP_comparison['RP_x'].quantile(.75)) & (RP_comparison['RP_x'] >= RP_comparison['RP_x'].quantile(.5)), '50-75%',
    np.where((RP_comparison['RP_x'] < RP_comparison['RP_x'].quantile(.90)) & (RP_comparison['RP_x'] >= RP_comparison['RP_x'].quantile(.75)), '75-90%', '90%>')))))
```

```python
pitchers = pitching_stats(2022, qual = 5)[["Name","WAR", "ERA", "K/9", "BB/9", "WHIP", "FIP", 'SIERA', 'Hard%', 'Soft%']]
```

```python
pitchers[['first', 'last']] = pitchers['Name'].str.split(' ', n= 1, expand= True)
pitchers['player_name'] = pitchers['last'] + ', ' + pitchers['first']
pitchers.drop(columns = ['Name', 'first', 'last'], inplace = True)
pitchers
```

```python
RP_stats = pd.merge(RP_comparison, pitchers, left_on = "player_name", right_on = "player_name")
```
