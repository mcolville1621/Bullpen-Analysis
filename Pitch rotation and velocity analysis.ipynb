```python
import pandas as pd
import numpy as np
from pybaseball import pitching_stats
```

```python
pitchers = pitching_stats(2022, qual = 30)
pitchers['Num_pitches'] = 7- pitchers[['FB% 2','SL%','CT%','CB%','CH%','SF%','KN%']].isnull().sum(axis=1)
stats = pitchers[["IP", "WAR", "ERA", "K/9", "BB/9", "WHIP", "FIP", 'SIERA', 'Hard%', 'Soft%', 'CStr%', 'CSW%']]
pitches = pitchers[['FB% 2','SL%','CT%','CB%','CH%','SF%']]
velo = pitchers[['FBv', 'SLv', 'CTv', 'CBv', 'CHv', 'SFv']]
FB_v = pitchers[['FBv',"WAR", "ERA", "SV", "K/9", "BB/9", "WHIP", "FIP", 'SIERA', 'Hard%', 'Soft%', 'CStr%', 'CSW%']]
SL_v = pitchers[['SLv',"WAR", "ERA", "SV", "K/9", "BB/9", "WHIP", "FIP", 'SIERA', 'Hard%', 'Soft%', 'CStr%', 'CSW%']]
CT_v = pitchers[['CTv',"WAR", "ERA", "SV", "K/9", "BB/9", "WHIP", "FIP", 'SIERA', 'Hard%', 'Soft%', 'CStr%', 'CSW%']]
CB_v = pitchers[['CBv',"WAR", "ERA", "SV", "K/9", "BB/9", "WHIP", "FIP", 'SIERA', 'Hard%', 'Soft%', 'CStr%', 'CSW%']]
CH_v = pitchers[['CHv',"WAR", "ERA", "SV", "K/9", "BB/9", "WHIP", "FIP", 'SIERA', 'Hard%', 'Soft%', 'CStr%', 'CSW%']]
SF_v = pitchers[['SFv',"WAR", "ERA", "SV", "K/9", "BB/9", "WHIP", "FIP", 'SIERA', 'Hard%', 'Soft%', 'CStr%', 'CSW%']]
```

```python
df['FB']= np.where(df['FB% 2'].isnull(), 0, 1)
df['SL'] = np.where(df['SL%'].isnull(), 0, 1)
df['CT'] = np.where(df['CT%'].isnull(), 0, 1)
df['CB'] = np.where(df['CB%'].isnull(), 0, 1)
df['CH'] = np.where(df['CH%'].isnull(), 0, 1)
df['SF']= np.where(df['SF%'].isnull(), 0, 1)
df_final = pd.merge(df, stats, left_index=True, right_index=True)
```

```python
stats_final = df_final.groupby(['FB', 'SL', 'CH', 'CB', 'SF', 'CT' ]).mean().drop(columns = ['FB% 2','SL%','CT%','CB%','CH%','SF%'])
Final = stats_final.reset_index()
Final['whiff rank'] = Final[['CStr%', 'CSW%']].apply(tuple,axis=1)\
             .rank(ascending=False).astype(int)
Final['WAR rank'] = Final[['WAR']].apply(tuple,axis=1)\
             .rank(ascending=False).astype(int)
Final['Hard contact rank'] = Final[['Hard%']].apply(tuple,axis=1)\
             .rank(ascending=True).astype(int)
Final['Soft contact rank'] = Final[['Soft%']].apply(tuple,axis=1)\
             .rank(ascending=False).astype(int)
Final['Contact rank'] = (Final['Hard contact rank'] + Final['Soft contact rank'])/2
Final['K rank'] = Final[['K/9']].apply(tuple,axis=1)\
             .rank(ascending=False).astype(int)
Final['Basic Pitching rank'] = Final[['ERA', 'WHIP', 'FIP']].apply(tuple,axis=1)\
             .rank(ascending=True).astype(int)
Final['overall rank'] = (Final['Contact rank'] + Final['K rank'] + Final['Basic Pitching rank'] + Final['whiff rank'])/4
Final.sort_values('overall rank')
```

# RP vs SP comparison focusing on 2 most dominant pitches because some RP only use two pitches

```python
pitchers1 = pitchers
```

```python
SP = pitchers1[pitchers1['Start-IP'] > 20]
RP = pitchers1[pitchers1['Relief-IP'] > 20]
```

```python
SP1 = SP[['FB% 2','SL%','CT%','CB%','CH%','SF%']]
RP1 = RP[['FB% 2','SL%','CT%','CB%','CH%','SF%']]
```

```python
arr = np.argsort(-SP1.values, axis=1)
df_SP = pd.DataFrame(SP1.columns[arr], index=SP1.index)
df_SP=df_SP[[0,1]]
df_SP['Type'] = 'SP'
```

```python
arr = np.argsort(-RP1.values, axis=1)
df_RP = pd.DataFrame(RP1.columns[arr], index=RP1.index)
df_RP = df_RP[[0,1]]
df_RP['Type'] = 'RP'
```

```python
SP_final = pd.merge(df_SP, SP, left_index = True, right_index = True)
RP_final = pd.merge(df_RP, RP, left_index = True, right_index = True)
SP_RP = pd.concat([SP_final, RP_final])
SP_RP['FB_b']= np.where((SP_RP[0] == 'FB% 2') | (SP_RP[1] == 'FB% 2'), 1, 0)
SP_RP['SL'] = np.where((SP_RP[0] == 'SL%') | (SP_RP[1] == 'SL%'), 1, 0)
SP_RP['CT'] = np.where((SP_RP[0] == 'CT%') | (SP_RP[1] == 'CT%'), 1, 0)
SP_RP['CB'] = np.where((SP_RP[0] == 'CB%') | (SP_RP[1] == 'CB%'), 1, 0)
SP_RP['CH'] = np.where((SP_RP[0] == 'CH%') | (SP_RP[1] == 'CH%'), 1, 0)
SP_RP['SF']= np.where((SP_RP[0] == 'SF%') | (SP_RP[1] == 'SF%'), 1, 0)
SP_RP = SP_RP[['FB_b', 'SL', 'CT', 'CB', 'CH', 'SF', 'Type', "WAR", "ERA","K/9", "BB/9", "WHIP", "FIP", 'SIERA', 'Hard%', 'Soft%', 'CStr%', 'CSW%']]
compare_types = SP_RP.groupby(['FB_b', 'SL', 'CT', 'CB', 'CH', 'SF','Type']).mean()
compare_types = compare_types.reset_index()
```

```python
compare_types['whiff rank'] = compare_types[['CStr%', 'CSW%']].apply(tuple,axis=1)\
             .rank(ascending=False).astype(int)
compare_types['Hard contact rank'] = compare_types[['Hard%']].apply(tuple,axis=1)\
             .rank(ascending=True).astype(int)
compare_types['Soft contact rank'] = compare_types[['Soft%']].apply(tuple,axis=1)\
             .rank(ascending=False).astype(int)
compare_types['Contact rank'] = (compare_types['Hard contact rank'] + compare_types['Soft contact rank'])/2
compare_types['Basic Pitching rank'] = compare_types[['ERA', 'WHIP', 'FIP']].apply(tuple,axis=1)\
             .rank(ascending=True).astype(int)
compare_types['K rank'] = compare_types[['K/9']].apply(tuple,axis=1)\
             .rank(ascending=False).astype(int)
compare_types['overall rank'] = (compare_types['Contact rank'] + compare_types['K rank'] + compare_types['Basic Pitching rank'] + compare_types['whiff rank'])/4
Pivot_types = compare_types.sort_values('FB_b', ascending = False).drop(columns = ['Soft contact rank', 'Hard contact rank', 'WAR', 'BB/9', 'SIERA'])
```
