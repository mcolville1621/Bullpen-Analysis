```python
HL = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\Splits Leaderboard Data 8th Inning 9th Inning High Leverage.csv")
HL['OPS'] = HL['SLG'] + HL['OBP']
HL1 = HL[["Tm","ERA", "HR", "wOBA", "OPS"]]
HL1 = HL1.add_suffix('_High_Lev')
```

```python
LL = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\Splits Leaderboard Data 8th Inning 9th Inning Low Leverage.csv")
LL['OPS'] = LL['SLG'] + LL['OBP']
LL1 = LL[["Tm","ERA", "HR", "wOBA", "OPS"]]
LL1 = LL1.add_suffix('_Low_Lev')
```

```python
ML = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\Splits Leaderboard Data 8th Inning 9th Inning Medium Leverage.csv")
ML['OPS'] = ML['SLG'] + ML['OBP']
ML1 = ML[["Tm","ERA", "HR", "wOBA", "OPS"]]
ML1 = ML1.add_suffix('_Med_Lev')
```

```python
EOG = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\Splits Leaderboard Data 8th Inning 9th Inning.csv")
EOG['OPS'] = EOG['SLG'] + EOG['OBP']
EOG1 = EOG[["Tm","ERA", "HR", "wOBA", "OPS"]]
EOG1 = EOG1.add_suffix('_end_gm')
```

```python
Merge = ov_tot.merge(HL1, left_on = "Tm_overall", right_on = "Tm_High_Lev")
Merge2 = Merge.merge(ML1, left_on = "Tm_overall", right_on = "Tm_Med_Lev")
Merge3 = Merge2.merge(LL1, left_on = "Tm_overall", right_on = "Tm_Low_Lev")
Final = Merge3.merge(EOG1, left_on = "Tm_overall", right_on = "Tm_end_gm")
Final = Final[['Tm_overall', 'ERA_overall', 'HR_overall', 'wOBA_overall',
       'OPS_overall', 'ERA_High_Lev', 'HR_High_Lev',
       'wOBA_High_Lev', 'OPS_High_Lev', 'ERA_Low_Lev',
       'HR_Low_Lev', 'wOBA_Low_Lev', 'OPS_Low_Lev',
       'ERA_Med_Lev', 'HR_Med_Lev', 'wOBA_Med_Lev', 'OPS_Med_Lev',
       'ERA_end_gm', 'HR_end_gm', 'wOBA_end_gm', 'OPS_end_gm']]
Final = Final.set_index('Tm_overall')
```

```python
HR = [col for col in Final.columns if 'HR' in col]
ERA = [col for col in Final.columns if 'ERA' in col]
wOBA = [col for col in Final.columns if 'wOBA' in col]
OPS = [col for col in Final.columns if 'OPS' in col]
Final[ERA].sort_values('ERA_overall', ascending= False)
```

```python
import glob
import os
all_files = glob.glob(os.path.join(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\ERA by inning\*.csv"))

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df['filename'] = os.path.basename(filename)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame['Inning'] = frame['filename'].str.partition(".")[0]
frame['OPS'] = frame['SLG'] + frame['OBP']
Combined_overall = frame[['Tm', 'G', 'ERA', 'HR', 'SO', 'wOBA', 'OPS', 'Inning']]
```

```python
MLB_teams = pd.read_excel(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\MLB Mapping file.xlsx")
```

```python
Inning_total = Combined_overall.merge(MLB_teams, left_on = "Tm", right_on = "Team_ID")
Inning_total.to_excel(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\Inning_overall.xlsx")
Inning_total['Inning'] = Inning_total['Inning'].astype(str).astype(int)
```
