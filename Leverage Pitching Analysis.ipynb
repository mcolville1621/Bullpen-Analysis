# Bullpen analysis Specifically in high leverage situations

```python
import glob
import pandas as pd
import os
import numpy as np
```

```python
HL_files = glob.glob(os.path.join(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\HL\*.csv"))

li1 = []

for filename in HL_files:
    df1 = pd.read_csv(filename, index_col=None, header=0)
    df1['filename'] = os.path.basename(filename)
    li1.append(df1)

frame1 = pd.concat(li1, axis=0, ignore_index=True)
```

## High leverage - Medium - Low leverage analysis by inning

```python
frame1['Inning'] = frame1['filename'].str.partition(".")[0]
frame1['OPS'] = frame1['SLG'] + frame1['OBP']
HL_inn = frame1[['Tm', 'ERA', 'OPS', 'Inning']]
HL_inn
```

```python
LL_files = glob.glob(os.path.join(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\LL\*.csv"))

li2 = []

for filename in LL_files:
    df2 = pd.read_csv(filename, index_col=None, header=0)
    df2['filename'] = os.path.basename(filename)
    li2.append(df2)

frame2 = pd.concat(li2, axis=0, ignore_index=True)

frame2['Inning'] = frame2['filename'].str.partition(".")[0]
frame2['OPS'] = frame2['SLG'] + frame2['OBP']
LL_inn = frame2[['Tm', 'ERA', 'OPS', 'Inning']]
LL_inn
```

```python
ML_files = glob.glob(os.path.join(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\ML\*.csv"))

li3 = []

for filename in ML_files:
    df3 = pd.read_csv(filename, index_col=None, header=0)
    df3['filename'] = os.path.basename(filename)
    li3.append(df3)

frame3 = pd.concat(li3, axis=0, ignore_index=True)

frame3['Inning'] = frame3['filename'].str.partition(".")[0]
frame3['OPS'] = frame3['SLG'] + frame3['OBP']
ML_inn = frame3[['Tm', 'ERA', 'OPS', 'Inning']]
ML_inn
```

```python
HL_ML = HL_inn.merge(ML_inn, left_on=['Tm', "Inning"], right_on= ['Tm', "Inning"])
HL_ML = HL_ML.rename(columns = {"ERA_x":"ERA_HL", "ERA_y": "ERA_ML", "OPS_x":"OPS_HL", "OPS_y":"OPS_ML"})
Final = HL_ML.merge(LL_inn, left_on=['Tm', "Inning"], right_on= ['Tm', "Inning"])
Final = Final.rename(columns = {"ERA":"ERA_LL", "OPS":"OPS_LL"})
Final.to_excel(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\Leverage_Data.xlsx")
Final
```

### SP vs RP analysis

```python
SP_RP_files = glob.glob(os.path.join(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\SP-RP\*.csv"))

li4 = []

for filename in SP_RP_files:
    df4 = pd.read_csv(filename, index_col=None, header=0)
    df4['filename'] = os.path.basename(filename)
    li4.append(df4)

frame4 = pd.concat(li4, axis=0, ignore_index=True)
frame4
```

```python
frame4['Inning'] = np.where(frame4['filename'].str.contains('5th Inning'), 5, np.where(frame4['filename'].str.contains('6th Inning'),6,7))
frame4['Leverage_Type'] = np.where(frame4['filename'].str.contains('Medium'), "Medium", "High")
frame4['Position_Type'] = np.where(frame4['filename'].str.contains('SP'), "SP", "RP")
SP_RP_comparison = frame4[["Tm", "ERA", "Inning", "Leverage_Type", "Position_Type"]]
```

```python
SP_RP_comparison.to_excel(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\SP_RP.xlsx")
```

```python
Beginning_files = glob.glob(os.path.join(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\Leverage beg\*.csv"))

lis = []

for filename in Beginning_files:
    df5 = pd.read_csv(filename, index_col=None, header=0)
    df5['filename'] = os.path.basename(filename)
    lis.append(df5)

frame5 = pd.concat(lis, axis=0, ignore_index=True)
frame5['Position_Type'] = np.where(frame5['filename'].str.contains('SP'), "SP", "RP")
ERA = frame5[['League', 'G','ERA', 'Position_Type']]
IP = frame5[['League', 'IP', 'Position_Type']]
```

```python
MLB = ERA.merge(IP, left_on = ['League', 'Position_Type'], right_on = ['League', 'Position_Type'])
MLB = MLB.dropna()
MLB.to_excel(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Bullpen Breakdown\MLB_High.xlsx")
```
